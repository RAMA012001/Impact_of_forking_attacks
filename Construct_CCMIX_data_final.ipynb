```python
from datetime import datetime
def data_preparation(data):
  data.set_index('time', inplace=True)
  data.index = pd.to_datetime(data.index)
```

```python
data_preparation(BTGm)
data_preparation(ETCm)
data_preparation(XVGm)
data_preparation(XZCm)
data_preparation(ZENm)
data_preparation(ADAm)
data_preparation(EOSm)
data_preparation(LTCm)
data_preparation(TRXm)
data_preparation(XLMm)
data_preparation(XMRm)
data_preparation(XRPm)
```

```python

#CCMIX data base construction
CCMIX=ADAm
CCMIX=pd.merge(CCMIX,EOSm.iloc[:,:2],left_index=True, right_index=True)
CCMIX=pd.merge(CCMIX,LTCm.iloc[:,:2],left_index=True, right_index=True)
CCMIX=pd.merge(CCMIX,TRXm.iloc[:,:2],left_index=True, right_index=True)
CCMIX=pd.merge(CCMIX,XLMm.iloc[:,:2],left_index=True, right_index=True)
CCMIX=pd.merge(CCMIX,XMRm.iloc[:,:2],left_index=True, right_index=True)
CCMIX=pd.merge(CCMIX,XRPm.iloc[:,:2],left_index=True, right_index=True)
CCMIX=pd.merge(CCMIX,ETCm[['close_ETHUSD','volume_ETHUSD']],left_index=True, right_index=True)
```

```python
#CCMIX construction
#defining the 90 day rolling marketcap
CCMIX['mcp_BTCUSD']=CCMIX['close_BTCUSD']*CCMIX['volume_BTCUSD']
CCMIX['rolling_mcp_BTCUSD'] = CCMIX['mcp_BTCUSD'].rolling(window=90, min_periods=1).sum()
CCMIX['mcp_ETHUSD']=CCMIX['close_ETHUSD']*CCMIX['volume_ETHUSD']
CCMIX['rolling_mcp_ETHUSD'] = CCMIX['mcp_ETHUSD'].rolling(window=90, min_periods=1).sum()
CCMIX['mcp_ADAUSD']=CCMIX['close_ADAUSD']*CCMIX['volume_ADAUSD']
CCMIX['rolling_mcp_ADAUSD'] = CCMIX['mcp_ADAUSD'].rolling(window=90, min_periods=1).sum()
CCMIX['mcp_EOSUSD']=CCMIX['close_EOSUSD']*CCMIX['volume_EOSUSD']
CCMIX['rolling_mcp_EOSUSD'] = CCMIX['mcp_EOSUSD'].rolling(window=90, min_periods=1).sum()
CCMIX['mcp_TRXUSD']=CCMIX['close_TRXUSD']*CCMIX['volume_TRXUSD']
CCMIX['rolling_mcp_TRXUSD'] = CCMIX['mcp_TRXUSD'].rolling(window=90, min_periods=1).sum()
CCMIX['mcp_XRPUSD']=CCMIX['close_XRPUSD']*CCMIX['volume_XRPUSD']
CCMIX['rolling_mcp_XRPUSD'] = CCMIX['mcp_XRPUSD'].rolling(window=90, min_periods=1).sum()
CCMIX['mcp_XLMUSD']=CCMIX['close_XLMUSD']*CCMIX['volume_XLMUSD']
CCMIX['rolling_mcp_XLMUSD'] = CCMIX['mcp_XLMUSD'].rolling(window=90, min_periods=1).sum()
CCMIX['mcp_LTCUSD']=CCMIX['close_LTCUSD']*CCMIX['volume_LTCUSD']
CCMIX['rolling_mcp_LTCUSD'] = CCMIX['mcp_LTCUSD'].rolling(window=90, min_periods=1).sum()
CCMIX['mcp_XMRUSD']=CCMIX['close_XMRUSD']*CCMIX['volume_XMRUSD']
CCMIX['rolling_mcp_XMRUSD'] = CCMIX['mcp_XMRUSD'].rolling(window=90, min_periods=1).sum()
```

```python

#defining the CCMIX index as the rolling marketcap weighted average of the closing prices of the composing cryptocurrencies
total_rolling_mcp=CCMIX['rolling_mcp_BTCUSD']+CCMIX['rolling_mcp_ETHUSD']+CCMIX['rolling_mcp_ADAUSD']+CCMIX['rolling_mcp_EOSUSD']+CCMIX['rolling_mcp_XLMUSD']+CCMIX['rolling_mcp_LTCUSD']+CCMIX['rolling_mcp_XRPUSD']+CCMIX['rolling_mcp_TRXUSD']
CCMIX['CCMIX']=(CCMIX['rolling_mcp_BTCUSD']*CCMIX['close_BTCUSD']+CCMIX['rolling_mcp_ETHUSD']*CCMIX['close_ETHUSD']+CCMIX['rolling_mcp_ADAUSD']*CCMIX['close_ADAUSD']+CCMIX['rolling_mcp_EOSUSD']*CCMIX['close_EOSUSD']+CCMIX['rolling_mcp_XLMUSD']*CCMIX['close_XLMUSD']+CCMIX['rolling_mcp_LTCUSD']*CCMIX['close_LTCUSD']+CCMIX['rolling_mcp_XRPUSD']*CCMIX['close_XRPUSD']+CCMIX['rolling_mcp_TRXUSD']*CCMIX['close_TRXUSD'])/total_rolling_mcp
```
