```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_returns_comparison(data, column='close_BTCUSD'):
    # Set up seaborn for prettier plots
    sns.set(style="whitegrid")

    # Calculate daily returns (non-compounded)
    data['Daily_Return'] = data[column].pct_change()

    # Calculate compounded returns
    data['Compounded_Return'] = (1 + data['Daily_Return']).cumprod()

    # Calculate non-compounded cumulative returns
    data['Non_Compounded_Return'] = data['Daily_Return'].cumsum()

    # Calculate log returns
    data['Log_Return'] = np.log(data[column] / data[column].shift(1))

    # Create the plot
    plt.figure(figsize=(12, 7))

    # Plot compounded returns
    plt.plot(data.index, data['Compounded_Return'], label="Compounded Returns", color='b', linewidth=2)

    # Plot non-compounded returns
    plt.plot(data.index, data['Non_Compounded_Return'], label="Non-Compounded Returns", color='r', linestyle='--', linewidth=2)

    # Plot log returns
    plt.plot(data.index, data['Log_Return'].cumsum(), label="Log Returns", color='g', linestyle='-.', linewidth=2)

    # Plot diff Compounded Returns and  Non_Compounded_Return
    plt.plot(data.index,data['Compounded_Return'] - data['Non_Compounded_Return'], label="Compounded_Return - Non_Compounded_Return", color='y', linewidth=2)

    # Plot diff Log_Return and  Non_Compounded_Return
    plt.plot(data.index,data['Log_Return'].cumsum() - data['Non_Compounded_Return'], label="Log_Return - Non_Compounded_Return", color='b', linewidth=2)

    # Title and labels
    plt.title('Compounded vs Non-Compounded vs Log Returns of BTC', fontsize=16)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Returns', fontsize=12)

    # Adding a legend
    plt.legend()

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45)

    # Show the plot
    plt.tight_layout()
    plt.show()

# Example usage
plot_returns_comparison(CCMIX)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_return_derivatives_subplots(data, column='close_BTCUSD'):
    # Set up seaborn for prettier plots
    sns.set(style="whitegrid")

    # Calculate daily returns (non-compounded)
    data['Daily_Return'] = data[column].pct_change()

    # Calculate compounded returns
    data['Compounded_Return'] = (1 + data['Daily_Return']).cumprod()

    # Calculate non-compounded cumulative returns
    data['Non_Compounded_Return'] = data['Daily_Return'].cumsum()

    # Calculate log returns
    data['Log_Return'] = np.log(data[column] / data[column].shift(1))

    # Calculate the derivatives using the diff() method
    data['Compounded_Return_Derivative'] = data['Compounded_Return'].diff()
    data['Non_Compounded_Return_Derivative'] = data['Non_Compounded_Return'].diff()
    data['Log_Return_Derivative'] = data['Log_Return'].diff()

    # Create subplots (3 rows, 1 column)
    fig, axes = plt.subplots(3, 1, figsize=(12, 12), sharex=True)

    # Plot derivative of compounded returns
    axes[0].plot(data.index, data['Compounded_Return_Derivative'], label="Derivative of Compounded Returns", color='b', linewidth=2)
    axes[0].set_title('Derivative of Compounded Returns', fontsize=14)
    axes[0].set_ylabel('Derivative', fontsize=12)
    axes[0].legend()

    # Plot derivative of non-compounded returns
    axes[1].plot(data.index, data['Non_Compounded_Return_Derivative'], label="Derivative of Non-Compounded Returns", color='r', linestyle='--', linewidth=2)
    axes[1].set_title('Derivative of Non-Compounded Returns', fontsize=14)
    axes[1].set_ylabel('Derivative', fontsize=12)
    axes[1].legend()

    # Plot derivative of log returns
    axes[2].plot(data.index, data['Log_Return_Derivative'], label="Derivative of Log Returns", color='g', linestyle='-.', linewidth=2)
    axes[2].set_title('Derivative of Log Returns', fontsize=14)
    axes[2].set_xlabel('Date', fontsize=12)
    axes[2].set_ylabel('Derivative', fontsize=12)
    axes[2].legend()

    # Rotate x-axis labels for better readability
    for ax in axes:
        plt.sca(ax)
        plt.xticks(rotation=45)

    # Adjust layout for spacing
    plt.tight_layout()
    plt.show()

# Example usage
plot_return_derivatives_subplots(CCMIX)
```

```python
#CCMIX construction
#defining the 90 day rolling marketcap
CCMIX['mcp_BTCUSD']=CCMIX['close_BTCUSD']*CCMIX['volume_BTCUSD']
CCMIX['rolling_mcp_BTCUSD'] = CCMIX['mcp_BTCUSD'].rolling(window=90, min_periods=1).sum()
CCMIX['mcp_ETHUSD']=CCMIX['close_ETHUSD']*CCMIX['volume_ETHUSD']
CCMIX['rolling_mcp_ETHUSD'] = CCMIX['mcp_ETHUSD'].rolling(window=90, min_periods=1).sum()
CCMIX['mcp_ADAUSD']=CCMIX['close_ADAUSD']*CCMIX['volume_ADAUSD']
CCMIX['rolling_mcp_ADAUSD'] = CCMIX['mcp_ADAUSD'].rolling(window=90, min_periods=1).sum()
CCMIX['mcp_EOSUSD']=CCMIX['close_EOSUSD']*CCMIX['volume_EOSUSD']
CCMIX['rolling_mcp_EOSUSD'] = CCMIX['mcp_EOSUSD'].rolling(window=90, min_periods=1).sum()
CCMIX['mcp_TRXUSD']=CCMIX['close_TRXUSD']*CCMIX['volume_TRXUSD']
CCMIX['rolling_mcp_TRXUSD'] = CCMIX['mcp_TRXUSD'].rolling(window=90, min_periods=1).sum()
CCMIX['mcp_XRPUSD']=CCMIX['close_XRPUSD']*CCMIX['volume_XRPUSD']
CCMIX['rolling_mcp_XRPUSD'] = CCMIX['mcp_XRPUSD'].rolling(window=90, min_periods=1).sum()
CCMIX['mcp_XLMUSD']=CCMIX['close_XLMUSD']*CCMIX['volume_XLMUSD']
CCMIX['rolling_mcp_XLMUSD'] = CCMIX['mcp_XLMUSD'].rolling(window=90, min_periods=1).sum()
CCMIX['mcp_LTCUSD']=CCMIX['close_LTCUSD']*CCMIX['volume_LTCUSD']
CCMIX['rolling_mcp_LTCUSD'] = CCMIX['mcp_LTCUSD'].rolling(window=90, min_periods=1).sum()
# CCMIX['mcp_XMRUSD']=CCMIX['close_XMRUSD']*CCMIX['volume_XMRUSD']
# CCMIX['rolling_mcp_XMRUSD'] = CCMIX['mcp_XMRUSD'].rolling(window=90, min_periods=1).sum()
```

```python
# defining the CCMIX index as the rolling marketcap weighted average of the closing prices of the composing cryptocurrencies
total_rolling_mcp=CCMIX['rolling_mcp_BTCUSD']+CCMIX['rolling_mcp_ETHUSD']+CCMIX['rolling_mcp_ADAUSD']+CCMIX['rolling_mcp_EOSUSD']+CCMIX['rolling_mcp_XLMUSD']+CCMIX['rolling_mcp_LTCUSD']+CCMIX['rolling_mcp_XRPUSD']+CCMIX['rolling_mcp_TRXUSD']
CCMIX['CCMIX']=(CCMIX['rolling_mcp_BTCUSD']*CCMIX['close_BTCUSD']+CCMIX['rolling_mcp_ETHUSD']*CCMIX['close_ETHUSD']+CCMIX['rolling_mcp_ADAUSD']*CCMIX['close_ADAUSD']+CCMIX['rolling_mcp_EOSUSD']*CCMIX['close_EOSUSD']+CCMIX['rolling_mcp_XLMUSD']*CCMIX['close_XLMUSD']+CCMIX['rolling_mcp_LTCUSD']*CCMIX['close_LTCUSD']+CCMIX['rolling_mcp_XRPUSD']*CCMIX['close_XRPUSD']+CCMIX['rolling_mcp_TRXUSD']*CCMIX['close_TRXUSD'])/total_rolling_mcp
```
